```python
g = 9.81
def f1(v0):
    n = len(v0)
    fig1,ax1 = plt.subplots()
    for i in range (n):
        t = np.linspace(0,2*v0[i]/g,1000)
        ax1.plot(t, v0[i]*t-0.5*g*t**2, label=v0[i])
        ax1.legend()
v=np.arange(0,3,0.5)
f1(v)
    
    
```

```python
import numpy as np
from scipy.optimize import fmin
import math
def y(t,v):
    return -(v*t-0.5*g*t**2)
t = np.linspace(0,2*v/g,1000)
for broj in v:
    rez = fmin(y(t,broj),np.array([0,0]))
    ax1.plot(rez[0],rez[1])
```

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
#Odredimo d iz jednadbe ravnine
phi_m = np.linspace(0, 2*np.pi)
phi_p = np.linspace(0, 2*np.pi)
X,Y = np.meshgrid(phi_m, phi_p)
Z = np.sin(np.sqrt(X**2+Y**2))
fig = plt.figure(figsize=(14,6))
ax = fig.add_subplot(1, 1, 1, projection='3d')
ax.plot_surface(X, Y, Z, rstride=5, cstride=4, linewidth=10,cmap='terrain')
plt.show()
```

```python
izraz = x**2+x+1
def kvadratna(izraz):
    a = izraz.coeff(x,2)
    b = izraz.coeff(x,1)
    c = izraz.coeff(x,0)
    x1  = (-b + sp.sqrt(b**2-4*a*c))/(2*a)
    x2  = (-b - sp.sqrt(b**2-4*a*c))/(2*a)
    return x1,x2
kvadratna(izraz)    
```
